```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0, 10, 100)

plt.plot(x, np.cos(x), label='cosine')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

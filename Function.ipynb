```python
def my_function(fname):
    print(fname + " " + "Deus")

my_function("Denis")
my_function("Stella")
my_function("Winfrida")
```

```python
def my_function(country = "Tanzania"):
    print("I am from" + " " + country)

my_function("Sweden")
my_function("India")
my_function()
my_function("Brazil")
```

```python
def my_function(food):
    for x in food:
        print(x)

fruits = ['apple', 'banana', 'cherry']

my_function(fruits)
```

```python
def my_function(x):
    return 5 * x

print(my_function(3))
print(my_function(5))
```

```python
def my_function(child3, child2, child1):
    print("The youngest child is " + child1)

my_function(child1 = 'Emil', child2 = 'John', child3 = 'Derick')
```

```python
def my_function(*kids):
    print("The youngest child is " + kids[1])

my_function('Emil', 'John', 'Derick')
```

```python
def tri_recursion(k):
    if (k > 0):
        result = k + tri_recursion(k - 1)
        print(result)
    else:
        result = 0
    return result

print("\n\nRecursion Example Results")

tri_recursion(6)
```

```python
x = lambda a : a + 10
print(x(5))
```

```python
x = lambda a, b : a * b
print(x(5, 6))
```

```python
x = lambda a, b, c : a + b + c
print(x(5, 6, 2))
```

```python
def myfunc(n):
    return lambda a : a * n

mydoubler = myfunc(3)

print(mydoubler(7))
```

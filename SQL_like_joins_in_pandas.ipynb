```python
import pandas as pd
```

```python
student_df = pd.DataFrame({
    'roll_no' : [102, 101, 104, 103, 105],
    'name' : ['Stella', 'Denis', 'Winfrida', 'Deus', 'Suzana'],
    'grade' : ['B', 'A', 'B', 'C', 'C'],
    'marks' : [15, 25, 15, 9, 8],
    'city' : ['Mtwara', 'Ubungo', 'Pwani', 'Temeke', 'Kigamboni']
})

student_df
```

```python
city_state_mapping = pd.DataFrame({
    'city' : ['Mtwara', 'Ubungo', 'Pwani', 'Temeke', 'Kigamboni'],
    'state' : ['Masasi', 'Msewe', 'Chalinze', 'Mbagala', 'Cheza']
})

city_state_mapping
```

```python
student_df.merge(city_state_mapping, how='left', on='city')
```

```python
roll_no = pd.DataFrame({
    'roll_no' : [102, 103]
})

roll_no
```

```python
student_df.merge(roll_no, how='right', on='roll_no')
```

```python
student_selection = pd.DataFrame({
    'roll_no' : [102, 105, 101],
    'company' : ['TLS', 'URAFIKI', 'AIRTEL'],
    'package(lpa)' : [8, 11, 14.5]
})

student_selection
```

```python
student_df
```

```python
student_df.merge(student_selection, how='outer')
```
